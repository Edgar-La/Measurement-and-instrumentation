# __Práctica de Medición e Instrumentación__

```
#Importar paquetes
import scipy.signal as sig
import matplotlib.pyplot as plt
```

```
#Programa 5-4
filt = sig.lti((25), (1, 4, 25))
plt.axhline(y=1, color='r', linestyle='-')
plt.plot(*filt.step())
plt.title('Respuesta a un impulso solitario'); plt.xlabel('Tiempo (seg)'); plt.ylabel('Amplitud');
plt.grid()
plt.show()
```

```
#Programa 5-8
filt = sig.lti((0, 1), (1, 0.2, 1))
plt.axhline(y=0, color='r', linestyle='-')
plt.plot(*filt.impulse())
plt.title('Respuesta a un impulso solitario'); plt.xlabel('Tiempo (seg)'); plt.ylabel('Amplitud');
plt.grid()
plt.show()
```

```
#Programa 5-9
filt = sig.lti((1, 0), (1, 0.2, 1))
plt.axhline(y=0, color='r', linestyle='-')
plt.plot(*filt.step())
plt.title('Respuesta a un impulso solitario'); plt.xlabel('Tiempo (seg)'); plt.ylabel('Amplitud');
plt.grid()
plt.show()
```
